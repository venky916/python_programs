```python
pas=input('Enter the passwoed')
def is_pass(pas):
  cond1=len(pas)>8
  upper=False
  lower=False
  punc=False
  for val in pas:
    if val.isupper():
      upper=True
      break
  for val in pas:
    if val.islower():
      lower=True
      break
  for val in pas:
    punct='''!"#$%&'()*+, -./:;<=>?@[\]^_`{|}~'''
    if val in punct:
      punc=True
      break

  return cond1 and upper and lower and punc

if is_pass(pas):
  print('Password is valid')
else:
  print("not an valid password")
```

```python
"""Strong number => sum of numbers =Factorial sum of numbers"""

num=int(input('Enter the Number'))

def fact(x):
  if x==0 or x==1:
    return 1
  else:
    return x*fact(x-1)

def strong_num(num):
  sum=0
  while num:
    val=num%10
    sum+=fact(val)
    num=num//10
  return sum

if num==strong_num(num):
  print('STRONG NUMBER')
else:
  print('not a strong Number')
```

```python
'''Perfect number -> 28=> 1+ 2+ 4+7+14==28'''

def perfect_num(num):
  sum=0
  for i in range(1,num):
    if num%i==0:
      sum+=i
  return sum

num=int(input('Enter the number'))

if num==perfect_num(num):
  print('Perfect number')
else:
  print('Not a Perfect number')
```

```python
'''Arm Strong Number 153=>1**3+5**3+3**3'''

num=int(input('Enter the number'))

def Arm_strong(num):
  sum=0
  while num:
    val=num%10
    sum+=val**3
    num//=10

  return sum

if num==Arm_strong(num):
  print('Arm Strong number')
else:
  print('Not a Arm strong number')
```

```python
'''GCD or HCF'''

a=int(input('Enter the number'))

b=int(input('Enter the number'))

factor=1
for i in range(2,min(a,b)+1):
  if a%i==0 and b%i==0:
    factor=i

print(factor)

#################################################################

def gcd(a,b):
  if a==0 or b==0:
    return a+b
  elif a==b:
    return a
  elif a>b:
    return gcd(b,a-b)
  else:
    return gcd(a,b-a)

print(gcd(a,b))
###########################################################################
def gcd1(a, b):

    # Find minimum of a and b
    result = min(a, b)

    while result:
        if a % result == 0 and b % result == 0:
            break
        result -= 1

    # Return the gcd of a and b
    return result

print(gcd1(a,b))
################################################################################

# Recursive function to return gcd of a and b
def gcd2(a,b):

    # Everything divides 0
    if (b == 0):
         return a
    return gcd(b, a%b)

print(gcd2(a,b))
```
